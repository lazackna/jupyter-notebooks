```python
import numpy as np

%matplotlib inline

import matplotlib.pyplot as plt

from sklearn.datasets import make_circles

```

```python
# Create a circle in an array.
X, y = make_circles(n_samples=1000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

```python
# Print X
X
```

```python
# Print the shape of X
X.shape
```

```python
# Make a figure of 5 by 5.
plt.figure(figsize=(5, 5))
# Plot first layer of the array as blue circles.
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
# Plot the second layer of the array as red x's
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
# Set limit of the plot.
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
# Create a legend.
plt.legend(['0', '1'])
# Give the window a title.
plt.title("Blue circles and Red crosses")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
# Make a sequential model.
model = Sequential()
```

```python
# Add a shape to the model.
model.add(Dense(4, input_shape=(2,), activation='tanh'))
```

```python
# Add a new shape to the model.
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model.
model.compile(SGD(learning_rate=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
# Try and fit eveything in the model.
model.fit(X, y, epochs=20)
```

```python
hticks = np.linspace(-1.5, 1.5, 101)
vticks = np.linspace(-1.5, 1.5, 101)
# Make a meshgrid which fits the hticks and vticks.
aa, bb = np.meshgrid(hticks, vticks)
# Concatenate the data in one long array.
ab = np.c_[aa.ravel(), bb.ravel()]
# Predict the samples.
c = model.predict(ab)
# Reshape the model with the new predicted shape.
cc = c.reshape(aa.shape)
```

```python
# Make a figure of 5 by 5.
plt.figure(figsize=(5, 5))
# Make a contour.
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
# Plot the data in blue circles and red x's.
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
# Set the limit of the plot.
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
# Create a legend.
plt.legend(['0', '1'])
# Give it a title.
plt.title("Blue circles and Red crosses")
```
